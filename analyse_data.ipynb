```python
df_anonymized = pd.DataFrame()
```

```python
# Convertir datenaissance en format datetime si ce n'est pas déjà fait
df_selected['datenaissance'] = pd.to_datetime(df_selected['datenaissance'], errors='coerce')

# Vérifier si la conversion a réussi
if pd.isnull(df_selected['datenaissance']).any():
    print("Il y a des valeurs invalides dans la colonne datenaissance.")

# Appliquer la généralisation sur l'âge
df_anonymized['date_naissance'] = pd.cut(df_selected['datenaissance'].dt.year, bins=[0, 1970, 1980, 1990, 2000, 2010], labels=['<1970', '1970-1980', '1980-1990', '1990-2000', '2000-2010'])

# Supprimer la colonne datenaissance d'origine
df_anonymized
```

```python
import secrets  # Import du module secrets

df_anonymized['tokenized_adresse'] = df_selected['adressepersonnelle_clesmediapost_codevoiehexacle'].apply(lambda x: secrets.token_hex(16))
df_anonymized
```

```python
def modify_name(name):
    if isinstance(name, str) and len(name) >= 4:
        modified_name = name[0] + 'X' + name[2] + 'X' + name[4:]  # Modifier le 1er et le 4ème caractère
        return modified_name
    return name

# Appliquer la fonction à la colonne 'nom'
df_anonymized['modified_nom'] = df_selected['nom'].apply(modify_name)
df_anonymized
```

```python
def modify_name(prenom):
    if isinstance(prenom, str) and len(prenom) >= 4:
        modified_name = 'Y' + prenom[1] + prenom[2] + 'Y' + prenom[4:]  # Modifier le 1er et le 4ème caractère
        return modified_name
    return prenom

# Appliquer la fonction à la colonne 'nom'
df_anonymized['modified_prenom'] = df_selected['prenom'].apply(modify_name)
df_anonymized
```

```python
def mask_email(email):
   if pd.isna(email) or not isinstance(email, str):
       return 'Nan'
   parts = email.split('@')
   username = parts[0]
   masked_username = '*' * len(username)
   masked_email = masked_username + '@' + parts[1]
   return masked_email

# Appliquer le masquage à la colonne 'email'
df_anonymized['email_masked'] = df_selected['coordonneespersonnelles_email'].apply(mask_email)

print(df_anonymized)
```

```python
import numpy as np
def add_noise_to_postal_code(code):
    try:
        noise = np.random.randint(-10, 10)  # Ajouter un nombre aléatoire entre -10 et 10
        noisy_code = int(code) + noise
        return str(noisy_code)
    except ValueError:  # En cas d'erreur de conversion
        return code  # Renvoie simplement le code postal original si la conversion échoue

# Appliquer la fonction add_noise_to_postal_code à la colonne 'adressepersonnelle_codepostal'
df_anonymized['adressepersonnelle_codepostal_noisy'] = df_selected['adressepersonnelle_codepostal'].apply(add_noise_to_postal_code)

print(df_anonymized)
```

```python
df_anonymized['adressepersonnelle_numerolibellevoie_clean'] = df_selected['adressepersonnelle_numerolibellevoie'].str.replace(r'^\d+\sRUE\s', '', regex=True)
df_anonymized['adressepersonnelle_numerolibellevoie_clean'] = df_anonymized['adressepersonnelle_numerolibellevoie_clean'].str.replace(r'\d+', '', regex=True)

df_anonymized
```

```python
import pandas as pd
import numpy as np



# Fonction pour anonymiser un numéro de téléphone en masquant sélectivement les chiffres
def mask_phone_number(phone_number):
    if isinstance(phone_number, str):  # Vérifie si la valeur est une chaîne de caractères
        anonymized_number = ''
        digits_masked = 0
        for char in phone_number:
            if char.isdigit():
                if digits_masked < 2:  # Remplace seulement les deux premiers chiffres
                    anonymized_number += np.random.choice(['XX', 'YY']) + ' '
                    digits_masked += 1
                else:
                    anonymized_number += char  # Conserve les autres chiffres
            else:
                anonymized_number += char  # Conserve les caractères non numériques
        return anonymized_number.strip()  # Retire les espaces en trop
    else:
        return phone_number  # Retourne la valeur d'origine si ce n'est pas une chaîne de caractères

# Appliquer la fonction à la colonne de numéros de téléphone
df_anonymized['coordonneespersonnelles_telephone_anonymized'] = df_selected['coordonneespersonnelles_telephone'].apply(mask_phone_number)

print(df_anonymized)
```

```python
import dash
from dash import dcc, html
import dash.dependencies as dd
import plotly.express as px
import pandas as pd

# Charger les données
df = df_anonymized


# Créer une application Dash
app = dash.Dash(__name__)

# Définir la mise en page de l'application
app.layout = html.Div([
    html.H1("Tableau de bord du dataset"),

    # Dropdown pour sélectionner une colonne
    dcc.Dropdown(
        id='dropdown-column',
        options=[{'label': col, 'value': col} for col in df.columns],
        value=df.columns[0]
    ),

    # Graphique interactif
    dcc.Graph(id='graph')
])

# Définir les callbacks pour rendre le graphique interactif
@app.callback(
    dd.Output('graph', 'figure'),
    [dd.Input('dropdown-column', 'value')]
)
def update_graph(selected_column):
    fig = px.histogram(df, x=selected_column, title=f'Distribution de {selected_column}')
    return fig

# Exécuter l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```
